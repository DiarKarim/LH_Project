## Load all data

```python
path = "D:/Projects/LH_Project/Face_Music_Expr_Analysis/Data_FaceRaw/"
files = os.listdir(path)

tmpDF = None
df = None

for file in files: 
    if "detailed" in file:
        tmpDF = pd.read_csv(path + file, sep='\t', skiprows=8)
        
        tmpDF.insert(0,"ParticipantID", file[0:13])
        tmpDF.insert(1,"Condition", file[14:24])
        
        if df is None:
            df = tmpDF
        else: 
            df = pd.concat((df,tmpDF))
```

### Compute Valence as happy - most negative value from the negative emotions

```python
def ValenceFunc(data_frame):
    valence = []
    for d in range(len(data_frame)):
        sad = data_frame['Sad'].iloc[d]
        angry = data_frame['Angry'].iloc[d]
        scared = data_frame['Scared'].iloc[d]
        disgust = data_frame['Disgusted'].iloc[d]
        maxNegative = np.max([sad, angry, scared, disgust])
        valence.append(data_frame['Happy'].iloc[d] - maxNegative)
    return pd.Series(valence, index=data_frame.index)
```

### Add Valence back into the dataframe (df)

```python
# df['Valence'] = df.groupby(['ParticipantID', 'Condition']).apply(lambda x: pd.Series(Valence(x))).values
df['Valence'] = df.groupby(['ParticipantID', 'Condition']).apply(ValenceFunc).reset_index(level=[0,1], drop=True)

df
```

```python
def time_to_seconds(data_frame):
    
    times = []
    
    for i in range(len(data_frame)):
    
        time_str = data_frame['Video Time'].values[i][-1-8:-1-3]
        # Split the string into hours and minutes
        minutes, seconds = time_str.split(':')

        # Convert hours and minutes to integers
        minutes = int(minutes)
        seconds = int(seconds)

        # Calculate total seconds
        total_seconds = (minutes * 60) + seconds
        
        times.append(total_seconds)

    return pd.Series(times, index=data_frame.index)
    # return total_seconds

# # Example usage
# time_str = "02:55"
# seconds = time_to_seconds(time_str)
# print(f"Time in seconds: {seconds}")
```

```python
def change_range(x, min_val = -0.89, max_val = 0.98):    
    # # Calculate the minimum and maximum values in x.
    # min_val = np.min(x)
    # max_val = np.max(x)

    # Calculate the range of values.
    range_val = max_val - min_val

    # Scale all values in x to be between 0 and 1.
    scaled_x = (x - min_val) / range_val

    # Shift all values in x to be between -1 and 1.
    shifted_x = scaled_x * 2 - 1

    return shifted_x
```
